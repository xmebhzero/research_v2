```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read csv file
df = pd.read_csv('./datasets/HousingData.csv')

# use sns.set() to create a nice gray background with white grid lines
sns.set()

# title for histogram
title = 'Median Boston Housing Prices'

# Adjust the figure size of your histogram using (horizontal, vertical) dimensions
plt.figure(figsize=(14,8))

# create histogram with transparency (alpha) to add 3D effect
plt.hist(df['MEDV'], color='green', alpha=0.6)

# set additional info title and labels
plt.title(title, fontsize=15)
plt.xlabel('1980 Median Value in Thousands')
plt.ylabel('Count')
plt.savefig(title, dpi=300)

# Show the histogram
plt.show()
```
